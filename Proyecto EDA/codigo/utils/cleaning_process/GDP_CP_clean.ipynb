```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.ticker as ticker
import pandas as pd
```

```python
data = pd.read_csv('../../data/dirty_data/GDP_CP.csv',sep = ';')
```

```python
data
```

```python
data.info()
```

```python
data = data.replace(':', pd.NaT)
```

```python
data = data.dropna()
```

```python
data = data.set_index("GDP current prices")
```

```python
for col in data.columns:
    data[col] = data[col].str.split(',', expand=True)[0].str.replace('\.', '')
```

```python
data.info()
```

```python
data = data.astype(int)
```

```python
data.to_csv('GDP_CP_clean.csv')
```
