```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('../../data/dirty_data/employment.csv',sep = ';', encoding='latin',decimal=",")
```

```python
data
```

```python
data.replace(':', pd.np.nan, inplace=True)
```

```python
data = data.set_index('Total employment(domestic concept)')
```

```python
for col in data.columns:
    data[col] = data[col].str.split(',', expand=True)[0].str.replace('\.', '')
```

```python
data = data.dropna(axis=0)
```

```python
data
```

```python
data= data.astype(int)
```

```python
data.info()
```

```python
data.to_csv('employment_clean.csv')
```
