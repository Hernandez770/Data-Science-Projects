```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/clean_data/Ingresos_fiscales_cat_clean.csv',sep = ';')
```

```python
data = data.set_index("Ingresos fiscales según categoria")
```

```python
data = data.replace(',', '.', regex=True).astype(float)
```

```python
data
```

```python
pie = data.plot(x='Ingresos fiscales según categoria', y='España 2019', kind='pie',legend=None,autopct='%1.1f%%');
pie.set_title('Ingresos fiscales según categoría en España 2019')
pie.set_ylabel('')
```

```python
datos_espana = [22.7, 6.0, 35.3, 7.1, 18.8, 6.1, 4.0]
categorias = ['Renta, benef. y gananc. capital\nsobre Personas físicas', 
              'Renta, benef. y gananc. capital\nsobre Personas jurídicas', 
              'Cotizaciones\nSeguridad\nSocial', 'Impuestos sobre\npropiedad', 
                'Impuestos Generales\ntipo IVA', 
              'Impuestos\nEspeciales', 'Otros']


fig, ax = plt.subplots(figsize=(10, 6))
ax.pie(datos_espana, radius=1, labels=categorias,autopct='%1.01f%%', pctdistance=0.8)

plt.title('Ingresos fiscales según categoría - España 2019', fontsize=14)

circulo_central = plt.Circle((0,0), 0.5, color='white')
fig.gca().add_artist(circulo_central)

plt.show()

```

```python
data_esp = data[data.columns[0:2]]

# Crear el gráfico circular
ax = data_esp.plot(x=data_esp.columns[0], y=data_esp.columns[1], kind='pie', legend=None, autopct='%1.1f%%', startangle=90)

# Agregar el título
ax.set_title('Ingresos fiscales según categoría en España 2019')

plt.show()
```
