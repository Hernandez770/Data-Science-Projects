```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.ticker as ticker
import pandas as pd
```

```python
data = pd.read_csv('../data/clean_data/HICP_clean.csv')
```

```python
data.head()
```

```python
data = data.set_index("HICP - Inflation rate")
```

```python
data = data.rename(index={"European Union (EU6-1958, EU9-1973, EU10-1981, EU12-1986, EU15-1995, EU25-2004, EU27-2007, EU28-2013, EU27-2020)": "European Union"})
```

```python
data = data.rename(index={"Germany (until 1990 former territory of the FRG)": "Germany"})
```

```python
data.head()
```

```python
# Seleccionar las filas y columnas que se desean graficar
data_to_plot = data.loc[["European Union", "Spain", "France", "Italy","Germany"], "2011":"2022"]

# Crear el gráfico de líneas
plt.plot(data_to_plot.T)
plt.legend(data_to_plot.index)
plt.xlabel('Año')
plt.ylabel('Tasa de Inflación (%)')
plt.title('Tasa de Inflación de European Union, Spain, France, e Italy')
plt.show()
```

```python

data_countries = data.loc[data.index.isin(countries)]
style.use('fivethirtyeight')

plt.figure(figsize=(12,8))
for country, color in zip(countries, colors):
    plt.plot(data.loc[country][:], label=country, linewidth=2.5, color=color)
plt.xlabel('Year')
plt.xticks(rotation=-45)
plt.ylabel('million €')
plt.title('GDP (current prices)')
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
#plt.legend()
plt.ticklabel_format(style='plain', axis='y')
plt.axhline(y = 0, color = 'black', linewidth = 1.4, alpha = .7)
#plt.ylim(bottom=750000, top=1350000)
#formatter = ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x).replace(',', '.'))
#plt.gca().yaxis.set_major_formatter(formatter)
plt.show
```
