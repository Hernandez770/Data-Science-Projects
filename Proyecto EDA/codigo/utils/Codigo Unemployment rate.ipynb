```python
# Set the countries and colors
countries = ['Spain', 'Germany']
colors = [[0,0,0],[255/255,0,128/255] , [86/255,180/255,233/255], [0,158/255,115/255], [213/255,94/255,0], [0,114/255,178/255], 
          [240/255,228/255,66/255], [110/255,190/255,60/255], 
          [200/255,100/255,200/255], [128/255,128/255,128/255], [0,255/255,128/255], [230/255,159/255,0],
          [128/255,0,255/255], [255/255,128/255,0], [0,128/255,255/255], [128/255,255/255,0], [0,200/255,255/255]]

# Create the figure
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data for each country
for i, country in enumerate(countries):
    # Extract the year and month from the column names
    years = [int(col.split('-')[0]) for col in data.columns[1:]]
    months = [int(col.split('-')[1]) for col in data.columns[1:]]
    
    # Extract the unemployment rate data for the current country
    rates = data.loc[country][1:].values
    
    # Plot the unemployment rates for each month and year
    for j, (year, month) in enumerate(zip(years, months)):
        ax.plot(j, rates[j], color=colors[i], marker='o')
    
    # Add a label for the country to the last plotted point
    ax.text(j+0.2, rates[j], country, fontsize=12, color=colors[i])
    
# Set the axis labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Unemployment rate %')
ax.set_title('Unemployment rate')

# Create the legend with only the years
#ax.legend(years, loc='upper left', bbox_to_anchor=(1.02, 1))

# Add a horizontal line at y=0
ax.axhline(y=0, color='black', linewidth=1.4, alpha=0.7)

# Rotate the x-axis labels for better readability
plt.xticks(range(len(years)), years, rotation=-45)

# Show the plot
plt.show()
```
