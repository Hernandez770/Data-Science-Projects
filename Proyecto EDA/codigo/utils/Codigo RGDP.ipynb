```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.ticker as ticker
import pandas as pd
```

```python
data = pd.read_csv('../data/clean_data/RGDP_growth_clean.csv')
```

```python
data
```

```python
data = data.set_index("Country")
```

```python
data = data.rename(index={"Germany (until 1990 former territory of the FRG)": "Germany"})
```

```python
#DISPLAY EUROPE
countries = ['European Union*','Spain', 'Germany', 'United Kingdom', 'France', 'Italy', 'Netherlands','Poland','Sweden']
colors = [[0,0,0],[255/255,0,128/255] , [86/255,180/255,233/255], [0,158/255,115/255], [213/255,94/255,0], [0,114/255,178/255], 
          [240/255,228/255,66/255], [110/255,190/255,60/255], 
          [200/255,100/255,200/255], [128/255,128/255,128/255], [0,255/255,128/255], [230/255,159/255,0],
          [128/255,0,255/255], [255/255,128/255,0], [0,128/255,255/255], [128/255,255/255,0], [0,200/255,255/255]]
```

```python
data_countries = data.loc[data.index.isin(countries)]
style.use('fivethirtyeight')

plt.figure(figsize=(12,8))
for country, color in zip(countries, colors):
    plt.plot(data.loc[country][:], label=country, linewidth=2.5, color=color)
plt.xlabel('Year')
plt.xticks(rotation=-45)
plt.ylabel('%')
plt.title('Real GDP growth')
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
#plt.legend()
plt.ticklabel_format(style='plain', axis='y')
plt.axhline(y = 0, color = 'black', linewidth = 1.4, alpha = .7)
plt.text(1.02, -0.1, "*from 2020 it does not include the UK", transform=plt.gca().transAxes, fontsize=10, ha='left')
#plt.ylim(bottom=750000, top=1350000)
#formatter = ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x).replace(',', '.'))
#plt.gca().yaxis.set_major_formatter(formatter)
plt.show
```
