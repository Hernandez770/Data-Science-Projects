```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import pandas as pd
```

```python
data = pd.read_csv('../data/clean_data/GGD_clean.csv')
```

```python
data = data.set_index("General Government Debt (Percent of GDP)")
```

```python
data
```

Cálculo de la media de los países seleccionados para el gráfico

```python
# Selecciona filas correspondientes a los países de interés
subset = data.loc[['Spain', 'Germany', 'United Kingdom', 'France', 'Italy', 'Netherlands', 'United States','China','Japan','India','Canada'], :]

# Calcula la media de cada columna
mean_row = subset.mean(axis=0)

# Agrega la nueva fila al DataFrame
data.loc['Mean of selected countries'] = mean_row
```

Gráfico DEBT %GDP SELECTED COUNTRIES WORLDWIDE

```python
countries = ['Mean of selected countries','Spain', 'Germany', 'United Kingdom', 'France', 'Italy', 'Netherlands', 'United States','China','Japan','India','Canada']
data_countries = data.loc[data.index.isin(countries)]

colors = [[0,0,0],[255/255,0,128/255] , [86/255,180/255,233/255], [0,158/255,115/255], [213/255,94/255,0], [0,114/255,178/255], 
          [240/255,228/255,66/255], [110/255,190/255,60/255], 
          [200/255,100/255,200/255], [128/255,128/255,128/255], [0,255/255,128/255], [230/255,159/255,0],
          [128/255,0,255/255], [255/255,128/255,0], [0,128/255,255/255], [128/255,255/255,0], [0,200/255,255/255]]
style.use('fivethirtyeight')

plt.figure(figsize=(12,8))
for country, color in zip(countries, colors):
    plt.plot(data.loc[country][1:], label=country, linewidth=2.5, color=color)
plt.xlabel('Year')
plt.xticks(rotation=-45)
plt.ylabel('Debt as % GDP')
plt.title('Central government debt over GDP')
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
plt.axhline(y = 0, color = 'black', linewidth = 1.4, alpha = .7)
plt.show

```
