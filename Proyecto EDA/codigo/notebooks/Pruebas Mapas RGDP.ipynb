```python
import matplotlib.pyplot as plt

# seleccionar los pases y aos de inters
data_to_plot = data.loc[["Poland", "Spain", "France", "Italy", "United Kingdom"], "2011":"2022"]

# crear una paleta de colores personalizada para los pases
colors = ["#FF5733", "#FFC300", "#DAF7A6", "#C70039", "#00ADB5"]

# crear el grfico de lneas con los colores personalizados y ancho de lnea
fig, ax = plt.subplots(figsize=(10,6))
for i, country in enumerate(data_to_plot.index):
    ax.plot(data_to_plot.loc[country], color=colors[i], linewidth=2, label=country)

# agregar ttulo y etiquetas
ax.set_title("Crecimiento del PIB en pases seleccionados de la UE (2011-2022)", fontsize=18)
ax.set_xlabel("Aos", fontsize=14)
ax.set_ylabel("Crecimiento real del PIB (%)", fontsize=14)

# agregar leyenda con el ttulo y tamao de fuente personalizados
ax.legend(fontsize=12, title="Pases", title_fontsize=14)

# quitar bordes y fondo de la grfica
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_linewidth(0.5)
ax.spines["left"].set_linewidth(0.5)
ax.set_facecolor("#F0F0F0")

# mostrar la grfica
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# seleccionar los datos del ao 2020
data_2020 = data[['Real GDP Growth', '2020']]

# crear un diccionario que asigne un color a cada pas en funcin de su tasa de crecimiento del PIB en el ao 2020
color_map = {}
for index, row in data_2020.iterrows():
    if row['2020'] < 0:
        color_map[row['Real GDP Growth']] = 'red'  # si la tasa de crecimiento es negativa, el color ser rojo
    else:
        color_map[row['Real GDP Growth']] = 'green'  # si la tasa de crecimiento es positiva, el color ser verde

# crear el mapa
map = Basemap(projection='merc', llcrnrlon=-20, llcrnrlat=30, urcrnrlon=60, urcrnrlat=75, resolution='l')
map.drawcoastlines()
map.drawcountries(linewidth=0.5)

# aadir el color de cada pas al mapa
for country_info in map.cntryr:
    country_name = country_info['name']
    if country_name in color_map:
        color = color_map[country_name]
        poly = country_info['polys'][0]
        x, y = zip(*poly)
        map.fill(x, y, color)

# aadir la leyenda
plt.title('Crecimiento del PIB en Europa en 2020')
red_patch = plt.Rectangle((0, 0), 1, 1, fc='red', edgecolor='black')
green_patch = plt.Rectangle((0, 0), 1, 1, fc='green', edgecolor='black')
plt.legend([red_patch, green_patch], ['Crecimiento negativo', 'Crecimiento positivo'], loc='upper left')

plt.show()
```
