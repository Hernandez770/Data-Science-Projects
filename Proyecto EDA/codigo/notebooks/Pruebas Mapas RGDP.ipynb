```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# seleccionar los datos del año 2020
data_2020 = data[['Real GDP Growth', '2020']]

# crear un diccionario que asigne un color a cada país en función de su tasa de crecimiento del PIB en el año 2020
color_map = {}
for index, row in data_2020.iterrows():
    if row['2020'] < 0:
        color_map[row['Real GDP Growth']] = 'red'  # si la tasa de crecimiento es negativa, el color será rojo
    else:
        color_map[row['Real GDP Growth']] = 'green'  # si la tasa de crecimiento es positiva, el color será verde

# crear el mapa
map = Basemap(projection='merc', llcrnrlon=-20, llcrnrlat=30, urcrnrlon=60, urcrnrlat=75, resolution='l')
map.drawcoastlines()
map.drawcountries(linewidth=0.5)

# añadir el color de cada país al mapa
for country_info in map.cntryr:
    country_name = country_info['name']
    if country_name in color_map:
        color = color_map[country_name]
        poly = country_info['polys'][0]
        x, y = zip(*poly)
        map.fill(x, y, color)

# añadir la leyenda
plt.title('Crecimiento del PIB en Europa en 2020')
red_patch = plt.Rectangle((0, 0), 1, 1, fc='red', edgecolor='black')
green_patch = plt.Rectangle((0, 0), 1, 1, fc='green', edgecolor='black')
plt.legend([red_patch, green_patch], ['Crecimiento negativo', 'Crecimiento positivo'], loc='upper left')

plt.show()
```
