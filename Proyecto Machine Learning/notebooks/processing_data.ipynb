## DATA MERGING

```python
import os

path = "Data merging/"
data_frames = []
```

```python
for file in os.listdir(path):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(path, file))
        data_frames.append(df)
```

```python
df = pd.concat(data_frames, ignore_index=True)
```

```python
df.insert(0, "ID", range(1, len(df)+1))
```

```python
df = df.set_index('ID')
```

```python
df
```

```python
df.to_csv("merged_20_years.csv", index=False)
```

```python
# ELIMINAR COLUMNAS CON DEMASIADOS NANS
df = df.dropna(thresh=len(df) - 1000, axis=1)
```

```python
# ELIMINAR ROWS CON DEMASIADOS NANS
df = df.dropna(thresh=len(df.columns) - 100, axis=0)
```

```python
df = df.drop(columns='Country Code')
```

```python
nan_counts = df.isna().sum()
for column_name, count in nan_counts.items():
    print(f"Column '{column_name}' has {count} NaN values.")
```
